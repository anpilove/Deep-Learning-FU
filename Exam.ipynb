#### Модель перцептрона

https://www.youtube.com/watch?v=t9QfcFNkG58&list=PLA0M1Bcd0w8yv0XGiF1wjerjSZVSrYbjh&index=3


Перцептрон — это простейший вид искусственной нейронной сети, предложенный Френком Розенблаттом в 1957 году. Он является основным строительным блоком для более сложных нейронных сетей. Перцептрон представляет собой модель, инспирированную биологическими нейронами, и используется в задачах бинарной классификации.

Основные компоненты перцептрона:

Входы (Input): Каждый вход представляет собой численное значение, которое можно интерпретировать как признак или характеристику входных данных.

Веса (Weights): Каждому входу присваивается вес, который отражает важность этого входа для выхода. Веса подбираются в процессе обучения.

Сумматор (Summation): Входные значения умножаются на соответствующие веса и суммируются.

Функция активации (Activation Function): Результат суммации передается через функцию активации, которая определяет, должен ли перцептрон активироваться (выдавать 1) или нет (выдавать 0).

Биас (Bias): Дополнительный параметр, который прибавляется к сумме взвешенных входов перед применением функции активации.

Простой пример использования перцептрона — решение логических задач типа "И" (AND) или "ИЛИ" (OR). В более сложных сценариях используются многослойные перцептроны (многослойные нейронные сети) для решения более сложных задач, таких как распознавание образов или классификация изображений.
